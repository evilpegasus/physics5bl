```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import pandas as pd

def st_err(datapts):
    return np.std(datapts, ddof=1)/np.sqrt(len(datapts))
def linear_model(x, m, c):
    return m*x + c

mu = 1.25663706212e-6
I = 3

a = open("data/one_wire/dataset-18e6ec07-73d9-48a7-881c-9d8ed9ce7260.csv", "r")
b = open("data/one_wire/dataset-26bc9bac-ca80-4b01-8591-5552013e2cd1.csv", "r")
c = open("data/one_wire/dataset-2b09a081-4bba-4744-8b8a-872458324185.csv", "r")
d = open("data/one_wire/dataset-31e8b80b-d9ad-410b-87ef-70f1c9d6850e.csv", "r")
e = open("data/one_wire/dataset-41075a01-5d24-4f00-9fc6-a4a2cd262c2f.csv", "r")
f = open("data/one_wire/dataset-519228f7-6691-40ce-94dd-a8cdd9d5b31a.csv", "r")
g = open("data/one_wire/dataset-7d7576a8-4c4f-437b-9c76-7a335549cbc8.csv", "r")
h = open("data/one_wire/dataset-8caeceaf-c72e-45cc-909f-6532a16e085d.csv", "r")
i = open("data/one_wire/dataset-bcdbb5c0-4eaf-45c1-8792-a1521eac5fe4.csv", "r")
j = open("data/one_wire/dataset-e98d7a03-91c4-45e9-bf2a-550788fc4058.csv", "r")
k = open("data/one_wire/dataset-f233f447-aff8-4f74-a068-e80b463d4082.csv", "r")

raw_data10 = np.loadtxt(j, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data13 = np.loadtxt(h, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data15 = np.loadtxt(k, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data17 = np.loadtxt(g, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data20 = np.loadtxt(b, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data25 = np.loadtxt(c, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data30 = np.loadtxt(d, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data35 = np.loadtxt(a, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data40 = np.loadtxt(f, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data45 = np.loadtxt(e, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)
raw_data50 = np.loadtxt(i, delimiter = ",", skiprows = 1, usecols=(0, 1, 2, 3), unpack = True)

range_off10 = (raw_data10[0] >= 0) & (raw_data10[0] <= 6)
range_on10 = (raw_data10[0] >= 7) & (raw_data10[0] <= 14)
range_off13 = (raw_data13[0] >= 0) & (raw_data13[0] <= 6)
range_on13 = (raw_data13[0] >= 7) & (raw_data13[0] <= 13)
range_off15 = (raw_data15[0] >= 0) & (raw_data15[0] <= 6)
range_on15 = (raw_data15[0] >= 7) & (raw_data15[0] <= 14)
range_off17 = (raw_data17[0] >= 0) & (raw_data17[0] <= 6)
range_on17 = (raw_data17[0] >= 7) & (raw_data17[0] <= 15)
range_off20 = (raw_data20[0] >= 0) & (raw_data20[0] <= 6)
range_on20 = (raw_data20[0] >= 7) & (raw_data20[0] <= 15)
range_off25 = (raw_data25[0] >= 0) & (raw_data25[0] <= 6)
range_on25 = (raw_data25[0] >= 7) & (raw_data25[0] <= 16)
range_off30 = (raw_data30[0] >= 0) & (raw_data30[0] <= 6)
range_on30 = (raw_data30[0] >= 7) & (raw_data30[0] <= 15)
range_off35 = (raw_data35[0] >= 0) & (raw_data35[0] <= 6)
range_on35 = (raw_data35[0] >= 7) & (raw_data35[0] <= 15)
range_off40 = (raw_data40[0] >= 0) & (raw_data40[0] <= 6)
range_on40 = (raw_data40[0] >= 8) & (raw_data40[0] <= 16)
range_off45 = (raw_data45[0] >= 0) & (raw_data45[0] <= 7)
range_on45 = (raw_data45[0] >= 8) & (raw_data45[0] <= 16)
range_off50 = (raw_data50[0] >= 0) & (raw_data50[0] <= 6)
range_on50 = (raw_data50[0] >= 8) & (raw_data50[0] <= 16)

B_off10 = np.mean(raw_data10[1][range_off10]/(10**6)), st_err(raw_data10[1][range_off10]/(10**6))
B_on10 = np.mean(raw_data10[1][range_on10]/(10**6)), st_err(raw_data10[1][range_on10]/(10**6))
B_off13 = np.mean(raw_data13[1][range_off13]/(10**6)), st_err(raw_data13[1][range_off13]/(10**6))
B_on13 = np.mean(raw_data13[1][range_on13]/(10**6)), st_err(raw_data13[1][range_on13]/(10**6))
B_off15 = np.mean(raw_data15[1][range_off15]/(10**6)), st_err(raw_data15[1][range_off15]/(10**6))
B_on15 = np.mean(raw_data15[1][range_on15]/(10**6)), st_err(raw_data15[1][range_on15]/(10**6))
B_off17 = np.mean(raw_data17[1][range_off17]/(10**6)), st_err(raw_data17[1][range_off17]/(10**6))
B_on17 = np.mean(raw_data17[1][range_on17]/(10**6)), st_err(raw_data17[1][range_on17]/(10**6))
B_off20 = np.mean(raw_data20[1][range_off20]/(10**6)), st_err(raw_data20[1][range_off20]/(10**6))
B_on20 = np.mean(raw_data20[1][range_on20]/(10**6)), st_err(raw_data20[1][range_on20]/(10**6))
B_off25 = np.mean(raw_data25[1][range_off25]/(10**6)), st_err(raw_data25[1][range_off25]/(10**6))
B_on25 = np.mean(raw_data25[1][range_on25]/(10**6)), st_err(raw_data25[1][range_on25]/(10**6))
B_off30 = np.mean(raw_data30[1][range_off30]/(10**6)), st_err(raw_data30[1][range_off30]/(10**6))
B_on30 = np.mean(raw_data30[1][range_on30]/(10**6)), st_err(raw_data30[1][range_on30]/(10**6))
B_off35 = np.mean(raw_data35[1][range_off35]/(10**6)), st_err(raw_data35[1][range_off35]/(10**6))
B_on35 = np.mean(raw_data35[1][range_on35]/(10**6)), st_err(raw_data35[1][range_on35]/(10**6))
B_off40 = np.mean(raw_data40[1][range_off40]/(10**6)), st_err(raw_data40[1][range_off40]/(10**6))
B_on40 = np.mean(raw_data40[1][range_on40]/(10**6)), st_err(raw_data40[1][range_on40]/(10**6))
B_off45 = np.mean(raw_data45[1][range_off45]/(10**6)), st_err(raw_data45[1][range_off45]/(10**6))
B_on45 = np.mean(raw_data45[1][range_on45]/(10**6)), st_err(raw_data45[1][range_on45]/(10**6))
B_off50 = np.mean(raw_data50[1][range_off50]/(10**6)), st_err(raw_data50[1][range_off50]/(10**6))
B_on50 = np.mean(raw_data50[1][range_on50]/(10**6)), st_err(raw_data50[1][range_on50]/(10**6))

delta_B10 = B_on10[0] - B_off10[0], np.sqrt(B_on10[1]**2 + B_off10[1]**2)
delta_B13 = B_on13[0] - B_off13[0], np.sqrt(B_on13[1]**2 + B_off13[1]**2)
delta_B15 = B_on15[0] - B_off15[0], np.sqrt(B_on15[1]**2 + B_off15[1]**2)
delta_B17 = B_on17[0] - B_off17[0], np.sqrt(B_on17[1]**2 + B_off17[1]**2)
delta_B20 = B_on20[0] - B_off20[0], np.sqrt(B_on20[1]**2 + B_off20[1]**2)
delta_B25 = B_on25[0] - B_off25[0], np.sqrt(B_on25[1]**2 + B_off25[1]**2)
delta_B30 = B_on30[0] - B_off30[0], np.sqrt(B_on30[1]**2 + B_off30[1]**2)
delta_B35 = B_on35[0] - B_off35[0], np.sqrt(B_on35[1]**2 + B_off35[1]**2)
delta_B40 = B_on40[0] - B_off40[0], np.sqrt(B_on40[1]**2 + B_off40[1]**2)
delta_B45 = B_on45[0] - B_off45[0], np.sqrt(B_on45[1]**2 + B_off45[1]**2)
delta_B50 = B_on50[0] - B_off50[0], np.sqrt(B_on50[1]**2 + B_off50[1]**2)

y10 = 1 / delta_B10[0], (1 / delta_B10[0])**2 * delta_B10[1]
y13 = 1 / delta_B13[0], (1 / delta_B13[0])**2 * delta_B13[1]
y15 = 1 / delta_B15[0], (1 / delta_B15[0])**2 * delta_B15[1]
y17 = 1 / delta_B17[0], (1 / delta_B17[0])**2 * delta_B17[1]
y20 = 1 / delta_B20[0], (1 / delta_B20[0])**2 * delta_B20[1]
y25 = 1 / delta_B25[0], (1 / delta_B25[0])**2 * delta_B25[1]
y30 = 1 / delta_B30[0], (1 / delta_B30[0])**2 * delta_B30[1]
y35 = 1 / delta_B35[0], (1 / delta_B35[0])**2 * delta_B35[1]
y40 = 1 / delta_B40[0], (1 / delta_B40[0])**2 * delta_B40[1]
y45 = 1 / delta_B45[0], (1 / delta_B45[0])**2 * delta_B45[1]
y50 = 1 / delta_B50[0], (1 / delta_B50[0])**2 * delta_B50[1]

radius = np.array([1.0, 1.3, 1.5, 1.7, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])*10**-2
y_values = abs(np.array([y10[0], y13[0], y15[0], y17[0], y20[0], y25[0], y30[0], y35[0], y40[0], y45[0], y50[0]]))
y_err = np.array([y10[1], y13[1], y15[1], y17[1], y20[1], y25[1], y30[1], y35[1], y40[1], y45[1], y50[1]])

opt_par, opt_cov = opt.curve_fit(linear_model, y_values, radius)
m_lin, c_lin = opt_par
dm_lin, dc_lin = np.sqrt(np.diag(opt_cov))

m_theory = mu*I/(2*np.pi)

print("m = %5.11f \u00b1 %5.11f " % (m_lin, dm_lin))
print("c = %5.8f \u00b1 %5.8f " % (c_lin, dc_lin))
print("Theoretical m should be m(the) = %5.11f" % (m_theory))

# plt.errorbar(y_values, radius, yerr = y_err, fmt = "ko", label = "data points")
plt.scatter(y_values, radius, label = "data points")
plt.plot(y_values, y_values * m_lin + c_lin, label = "linear fit")
plt.plot(y_values, y_values * m_theory - c_lin, label = "theoretical fit")
plt.ylabel("radius, r (m)")
plt.xlabel("1 / magnetic field, 1/|B| (T^-1)")
plt.legend()
plt.title("Inverse Magnetic Field vs Distance (All Points)")
plt.show()

def chisq(predicted, observed, errors):
    res = observed-predicted
    norm_res = res / errors 
    return np.sum(norm_res**2)

chisq1 = chisq(radius, y_values * m_lin + c_lin, [0.001]*11)
reduced_chisq1 = chisq1 / 9

print("reduced chisq =", reduced_chisq1)

opt_par2, opt_cov2 = opt.curve_fit(linear_model, y_values[0:6], radius[0:6])
m_lin2, c_lin2 = opt_par2
dm_lin2, dc_lin2 = np.sqrt(np.diag(opt_cov2))

m_theory = mu*I/(2*np.pi)

print("m = %5.11f \u00b1 %5.11f " % (m_lin2, dm_lin2))
print("c = %5.8f \u00b1 %5.8f " % (c_lin2, dc_lin2))
print("Theoretical m should be m(the) = %5.11f" % (m_theory))

plt.scatter(y_values[0:6], radius[0:6], label = "data points")
plt.plot(y_values[0:6], y_values[0:6] * m_lin2 + c_lin2, label = "fit")
plt.plot(y_values[0:6], y_values[0:6] * m_theory - c_lin2, label = "theoretical fit")
plt.ylabel("radius, r (m)")
plt.xlabel("1 / magnetic field, 1/|B| (T^-1)")
plt.legend()
plt.title("Inverse Magnetic Field vs Distance (First 6 Points)")
plt.show()

def chisq(predicted, observed, errors):
    res = observed-predicted
    norm_res = res / errors 
    return np.sum(norm_res**2)

chisq2 = chisq(radius[0:6], y_values[0:6] * m_lin2 + c_lin2, [0.0005]*6)
reduced_chisq2 = chisq2 / 4

print("reduced chisq 2 =", reduced_chisq2)
```
