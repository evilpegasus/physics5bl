```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("data\EM1 Lab Data - Sheet1.csv")
df
```

```python
df_2b = df[["2b ming voltage", "2b ming v std", "2b ming current", "2b ming i std (mV)"]].dropna()
df_2b
```

# 2b

```python
plt.errorbar(df_2b["2b ming voltage"], df_2b["2b ming current"], xerr=df_2b["2b ming v std"], yerr=df_2b["2b ming i std (mV)"], fmt='o')
plt.xlabel("DAC Voltage (V)")
plt.ylabel("Current (mA)")
plt.title("Current vs Voltage (Non-Ohmic LED)")
```

```python
df_2b.columns = ["Voltage (V)", "Voltage SE", "Current (mA)", "Current SE"]
df_2b
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(df_2b[["Voltage (V)"]], df_2b[["Current (mA)"]])
```

```python
plt.plot(df_2b["Voltage (V)"], lr.predict(df_2b[["Voltage (V)"]]), color='red')
plt.scatter(df_2b["Voltage (V)"], df_2b["Current (mA)"], color='blue')
plt.xlabel("DAC Voltage (V)")
plt.ylabel("Current (mA)")
plt.title("Current vs Voltage (Non-Ohmic Green LED)")
plt.show()
print("m = " + str(lr.coef_[0]) + " +/- " + str(lr.coef_[0] * 0.0005))
print("b = " + str(lr.intercept_) + " +/- " + str(lr.intercept_ * 0.0005))
```

```python
# plot residuals
plt.scatter(lr.predict(df_2b[["Voltage (V)"]]), lr.predict(df_2b[["Voltage (V)"]]) - df_2b[["Current (mA)"]], color='black')
plt.axhline(y=0, color='r', linestyle='-')
plt.xlabel("DAC Voltage (V)")
plt.ylabel("Predicted - Actual Current (mA)")
plt.title("Current vs Voltage Residuals (Non-Ohmic Green LED)")
```

```python
str((1 / lr.coef_[0])[0]) + "+/-" + str((1 / lr.coef_[0])[0] * 0.0005)
```
